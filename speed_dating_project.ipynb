```python
from scipy.stats import chi2_contingency
```

```python
# Création des tables de contingence pour hommes et femmes
field_contingency_man = pd.crosstab(df_male["field_cd"], df_male["match"])
field_contingency_female = pd.crosstab(df_female["field_cd"], df_female["match"])
career_contingency_man = pd.crosstab(df_male["career_c"], df_male["match"])
career_contingency_female = pd.crosstab(df_female["career_c"], df_female["match"])

# Test Field du Chi² pour les hommes
chi2_field_man, p_value_f_man, dof_f_man, expected_f_man = chi2_contingency(field_contingency_man)

# Test Field du Chi² pour les femmes
chi2_field_female, p_value_f_female, dof_f_female, expected_f_female = chi2_contingency(field_contingency_female)

# Test Career du Chi² pour les hommes
chi2_career_man, p_value_c_man, dof_c_man, expected_c_man = chi2_contingency(career_contingency_man)

# Test Career du Chi² pour les femmes
chi2_career_female, p_value_c_female, dof_c_female, expected_c_female = chi2_contingency(career_contingency_man)

# Affichage des résultats
print(f"p_value field Man: {p_value_f_man}")
print(f"p_value field Female: {p_value_f_female}")
print(f"p_value career Man: {p_value_c_man}")
print(f"p_value career Female: {p_value_c_female}")
```

```python
import scipy.stats as stats
```

```python
# Transformation de la statistique Chi² en loi normale standard pour les hommes et les femmes
z_score_f_man = (chi2_field_man - dof_f_man) / np.sqrt(2 * dof_f_man)
z_score_f_female = (chi2_field_female - dof_f_female) / np.sqrt(2 * dof_f_female)
z_score_c_man = (chi2_career_man - dof_c_man) / np.sqrt(2 * dof_c_man)
z_score_c_female = (chi2_career_female - dof_c_female) / np.sqrt(2 * dof_c_female)

# Calcul du seuil critique pour p-value = 0.05 dans une loi normale standard
critical_z = stats.norm.ppf(0.95)  # Valeur de Z correspondant à une p-value de 0.05

# Génération de la loi normale standard
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)  # Loi normale standard 𝒩(0,1)

# Création du graphique
plt.figure(figsize=(10, 6))
plt.plot(x, y, color="blue", label="Loi Normale Standard")

# Ajout des valeurs de Z observées pour les hommes et les femmes
plt.axvline(z_score_f_man, color="red", linestyle="--", label=f"Valeur Z field Man: {z_score_f_man:.2f}")
plt.axvline(z_score_f_female, color="green", linestyle="--", label=f"Valeur Z field Female: {z_score_f_female:.2f}")
plt.axvline(z_score_c_man, color="blue", linestyle="--", label=f"Valeur Z career Man: {z_score_c_man:.2f}")
plt.axvline(z_score_c_female, color="pink", linestyle="--", label=f"Valeur Z career Female: {z_score_c_female:.2f}")

# Ajout de la ligne pour la p-value à 0.05
plt.axvline(critical_z, color="white", linestyle="dotted", label=f"Seuil p=0.05 (Z={critical_z:.2f})")

# Remplissage de la zone au-delà de p=0.05
plt.fill_between(x, y, where=(x >= critical_z), color="white", alpha=0.3, label="Zone de rejet H₀ (p<0.05)")

# Labels et titre
plt.xlabel("Score Z", fontsize=12)
plt.ylabel("Densité de Probabilité", fontsize=12)
plt.title("Visualisation du Test Chi² via la Loi Normale", fontsize=14)
plt.legend()
plt.grid()

# Affichage du graphique
plt.show()
```

```python
from scipy.stats import pearsonr
# Nettoyage des données pour éviter les NaN et infs
df_filt_gender_36 = df_filt_gender_36.dropna(subset=attributes + ["match"])

# Initialisation des dictionnaires pour stocker les p-values
p_values_male = {}
p_values_female = {}

# Test de significativité des corrélations pour chaque attribut chez les hommes (notés par les femmes)
for attr in attributes:
    for age_group in labels:
        subset = df_filt_gender_36[(df_filt_gender_36["gender"] == 1) & (df_filt_gender_36["age_group"] == age_group)]
        if len(subset) > 10 and subset[attr].notna().sum() > 10:  # Vérification des NaN et taille suffisante
            r, p = pearsonr(subset[attr], subset["match"])
            p_values_male[(attr, age_group)] = p

# Test de significativité des corrélations pour chaque attribut chez les femmes (notées par les hommes)
for attr in attributes:
    for age_group in labels:
        subset = df_filt_gender_36[(df_filt_gender_36["gender"] == 0) & (df_filt_gender_36["age_group"] == age_group)]
        if len(subset) > 10 and subset[attr].notna().sum() > 10:  # Vérification des NaN et taille suffisante
            r, p = pearsonr(subset[attr], subset["match"])
            p_values_female[(attr, age_group)] = p

# Transformer le dictionnaire en DataFrame
p_values_male_df = pd.DataFrame(list(p_values_male.items()), columns=["Attribut_Tranche", "p-value"])

# Séparer la colonne contenant les tuples (Attribut, Tranche d'âge) en deux colonnes distinctes
p_values_male_df[["Attribut", "Tranche d'âge"]] = pd.DataFrame(p_values_male_df["Attribut_Tranche"].tolist(), index=p_values_male_df.index)

# Supprimer la colonne temporaire contenant les tuples
p_values_male_df = p_values_male_df.drop(columns=["Attribut_Tranche"])
```
